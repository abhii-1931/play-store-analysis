```python
dict = {}
for i in data:
    dict[i] = list(df[df['App'] == i].index)
```

```python
duplicates = []
for i in dict.values():
    for index, j in enumerate(i):
        if index == 0:
            k = j
        else:
            if all((df.iloc[k, :] == df.iloc[j, :]).values):
                duplicates.append(j)
            else:
                z = 0
                for i in (df.iloc[k, :] == df.iloc[j, :]).values:
                    if i:
                        z+=1
                    else:
                        pass
                if z==12:
                    duplicates.append(j)
```

### Genres

```python
list_genres = []
for i in main_df['Genres']:
    if ';' in i:
        i = i.split(';')[-1]
        for j in i.split('&'):
            list_genres.append(j)
    else:
        for j in i.split('&'):
            list_genres.append(j)
    
```

```python
from wordcloud import WordCloud

# Example text data
text_data = ''
for i in list_genres:
    text_data = text_data +" "+ i 

text_data = ' '.join(text_data.split())
```

```python
# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white', collocations=False).generate_from_text(text_data)


# Plot the WordCloud image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axis labels
plt.show()
```
